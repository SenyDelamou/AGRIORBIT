```python
print(df.info())
print(df.describe())
```

## Analyze Global Trends

### Subtask:
Calculate and analyze the global average life expectancy, population, and GDP per capita over the years to identify overall trends.


**Reasoning**:
To analyze global trends, I need to group the DataFrame by 'year' and then calculate the mean for 'lifeExp', 'pop', and 'gdpPercap' columns to get the global averages over time.



```python
global_trends_df = df.groupby('year')[['lifeExp', 'pop', 'gdpPercap']].mean().reset_index()
print("Global Trends DataFrame created successfully:")
print(global_trends_df.head())
```

**Reasoning**:
Now that the global trends have been calculated, the next step is to visualize these trends using line plots to clearly show the evolution of life expectancy, population, and GDP per capita over the years.



```python
fig_lifeExp = px.line(global_trends_df, x='year', y='lifeExp', title='Global Average Life Expectancy Over Years', markers=True)
fig_lifeExp.show()

fig_pop = px.line(global_trends_df, x='year', y='pop', title='Global Average Population Over Years', markers=True)
fig_pop.show()

fig_gdp = px.line(global_trends_df, x='year', y='gdpPercap', title='Global Average GDP Per Capita Over Years', markers=True)
fig_gdp.show()
```

## Analyze Continental Trends

### Subtask:
Group the data by 'continent' and 'year' to analyze the average life expectancy, population, and GDP per capita for each continent over time.


**Reasoning**:
To analyze continental trends, I will group the main DataFrame `df` by 'continent' and 'year' and then calculate the mean of 'lifeExp', 'pop', and 'gdpPercap' for each group. Finally, I will reset the index and store the result in `continental_trends_df`.



```python
continental_trends_df = df.groupby(['continent', 'year'])[['lifeExp', 'pop', 'gdpPercap']].mean().reset_index()
print("Continental Trends DataFrame created successfully:")
print(continental_trends_df.head())
```

**Reasoning**:
Now that the continental trends have been calculated, the next step is to visualize these trends using line plots to clearly show the evolution of life expectancy, population, and GDP per capita for each continent over the years. This will help in comparing trends across different continents.



```python
fig_lifeExp_continent = px.line(continental_trends_df, x='year', y='lifeExp', color='continent', title='Continental Average Life Expectancy Over Years', markers=True)
fig_lifeExp_continent.show()

fig_pop_continent = px.line(continental_trends_df, x='year', y='pop', color='continent', title='Continental Average Population Over Years', markers=True)
fig_pop_continent.show()

fig_gdp_continent = px.line(continental_trends_df, x='year', y='gdpPercap', color='continent', title='Continental Average GDP Per Capita Over Years', markers=True)
fig_gdp_continent.show()
```
